```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from datetime import datetime,timedelta
```

```python
class pers:
    def __init__(self,name,date):
        self.name=name
        self.date=date
```

```python
#plt.scatter([datetime(1990,1,22)],[4])
def drawPers(xywh,name,pb):
    x=xywh[0]
    y=xywh[1]
    w=xywh[2]
    h=xywh[3]
    pb.plot([x,x+w,x+w,x,x],[y,y,y+h,y+h,y])
    pb.text(x+w/2,y+h/2,name,ha="center",va="center")
#drawPers([datetime(1990,1,22),0,timedelta(days=4),4],"Mig")
#print(datetime(1992,1,22)+timedelta(1,1,22))
#datetime.now()
```

```python
def drawRoom(room,roomnum,pb):
    rlen=[]
    for i in range(len(room)):
        if i+1<len(room):
            rlen.append(room[i+1].date-room[i].date)
        else:
            rlen.append(datetime.now()-room[i].date)
    #print(rlen)
    for p in zip(room,rlen):
        drawPers([p[0].date,roomnum-1/2,p[1],1],p[0].name,pb)
```

```python
data=pd.read_csv("3_2_data.txt",delimiter = " ",header=None)
pa,pb=plt.subplots(1,1,figsize=(3*6.4,3/2*4.8))
pb.set_yticks(range(315,329))
pb.set_yticklabels(range(315,329))
pb.set_xlim(datetime(2016,6,1),datetime.now()+timedelta(weeks=3))
xlab=[]
#curxlab=datetime(2016,6,1)
#while curxlab<datetime.now():
#    xlab.append(curxlab)
#    curxlab+=timedelta(days=182.5)
for da in range(2016,2023):
    xlab.append(datetime(da,1,1))
    xlab.append(datetime(da,7,1))
pb.set_xticks(xlab)
#print(len(data.index))
curroom=0
myroom=[]
for i,r in data.iterrows():
    if curroom != r[0]:
        drawRoom(myroom,curroom,pb)
        myroom=[]
        curroom=r[0]
    myroom.append( pers(r[4],datetime(r[1],r[2],r[3])) )
drawRoom(myroom,curroom,pb)
```
