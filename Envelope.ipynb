```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N=200
x=np.linspace(.001,1,N)
sig=np.zeros(N)
sig[5]=.2
sig[93]=.3
sig[103]=.4
env=np.exp(-(x*10)**2)
```

```python
fig=plt.figure(figsize=(25,10))
plt.plot(x,env,label="Envelope")
plt.plot(x,sig,label="Signal")
```

```python
col=np.zeros(N)
for i in range(N):
    for j in range(N):
        if i-j>=0:
            col[i]+=sig[j]*env[i-j]
plt.plot(x,col)
```
