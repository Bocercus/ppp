# Beatles
$x_0=0=y_1$  
$y_0=1=1-x_1$  
$x_1=0=1-y_0$  
$y_1=0=x_0$  
$\frac{d}{dt}x_1(t)=x_0-x_1=y_1-x_1$  
$\frac{d}{dt}y_1(t)=y_0-y_1=1-x_1-y_1$  

```python
import scipy.integrate as scpi
```

```python
tsp=[0,30]
st=10000
mag=1
def ddt(t,y):
    return [mag*(y[1]-y[0]),mag*(1-y[0]-y[1])]
def ev(t,y):
    return ((y[0]-y[1])+(1-y[0]-y[1]))
ev.terminal=True
sol=scpi.solve_ivp(ddt,tsp,[0,0],t_eval=np.linspace(tsp[0],tsp[1],st))#,events=ev)
```

```python
mm=100
plt.scatter(sol.y[0,:mm],sol.y[1,:mm])
plt.scatter(sol.y[1,:mm],1-sol.y[0,:mm])
plt.scatter(1-sol.y[0,:mm],1-sol.y[1,:mm])
plt.scatter(1-sol.y[1,:mm],sol.y[0,:mm])
```

```python
bet_len=0
for c in range(st-1):
    bet_len+=np.sqrt((sol.y[0,c+1]-sol.y[0,c])**2+(sol.y[1,c+1]-sol.y[1,c])**2)
    print(c,bet_len)
#c=6
#(sol.y[0,c+1]-sol.y[0,c])**2
```
