```python
pts=[]
def app(n):
    for i in range(n):
        pts.append([r.random(),r.random()])
def getRat(m=0):
    n=0
    if m==0:
        ptsm=pts
    else:
        ptsm=pts[0:m]
    for i in ptsm:
        if(np.sqrt(i[0]**2+i[1]**2)<=1.0):
            n=n+1
    return n/len(pts)
```

```python
app(8000)
```
